```python

import numpy as np
from matplotlib import pyplot



a = np.arange(0, 3.1,0.3)  #0から2πまでの範囲を100分割したnumpy配列
y=15*a-5*a**2
x=26*a

pyplot.xlim(0,78) #x軸範囲指定
pyplot.ylim(0,12) #y軸範囲指定
print(y)
print(x)
pyplot.plot(26*a,15*a-5*a**2)
pyplot.show()
```
